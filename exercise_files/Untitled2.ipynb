```python
import csv
```

```python
with open('10_02_us.csv','r') as f:
    reader=csv.reader(f, delimiter='\t')
    ctr=1
    next(reader)
    for row in reader:
        print(row)
        ctr+=1
        if ctr >=100:
            break
```

```python
with open('10_02_us.csv','r') as f:
    reader=csv.DictReader(f, delimiter='\t')
    ctr=1
    next(reader)
    for row in reader:
        print(row)
        ctr+=1
        if ctr >=20:
            break
```

```python
with open('10_02_us.csv','r') as f:
    data=list(csv.DictReader(f, delimiter='\t'))
```

```python
primes=[2]
for number in range(3,99999):
    for factor in primes:
        if number % factor == 0 and factor <= int(number**0.5)+1 :
            break
    else:
        primes.append(number)
print(primes)
```

```python
data_filtered=[row for row in data if int(row['postal code']) in primes]
print(data_filtered)
```

```python
with open('10_02_us_k.csv','w') as f:
    writer=csv.writer(f)
    writer.writerow(['Postal Code','Place Name','County','State Code','State'])
    for row in data_filtered:
        writer.writerow([row['postal code'],row['place name'],row['county'],row['state code'],row['state']])
```

```python
import json
```

```python
jsonString='{"a":"apple","b":"banana","c":"cherry"}'
json.loads(jsonString)
```

```python
jsonString='{"a":"apple","b":"banana","c":"cherry",}'
json.loads(jsonString)
```

```python
jsonString='{"a":"apple","b":"banana","c":"cherry",}'
try:
    json.loads(jsonString)
except json.JSONDecodeError:
    print('Can\'t parse!')
```

```python
pythonDict={'a': 'apple', 'b': 'banana', 'c': 'cherry',}
```

```python
json.dumps(pythonDict)
```

```python
class Animal:
    def __init__(self,name):
        self.name=name
pythonDict={'a': Animal('antelope'), 'b':Animal('bear'), 'c': Animal('cat'),}
```

```python
json.dumps(pythonDict)
```

```python
class AnimalEncoder(json.JSONEncoder):
    def default(self,o):
        if type(o)==Animal:
            return o.name
        return super().default(o)
json.dumps(pythonDict, cls=AnimalEncoder)            
```
